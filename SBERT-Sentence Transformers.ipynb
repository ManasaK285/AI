```python
from sentence_transformers import SentenceTransformer,util
```

```python
model=SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences=['the cat sits outside','the new movie is awesome','the new movie is really great','the dog barks on strangers']
```

```python
embeddings=model.encode(sentences=sentences,convert_to_tensor=True)
```

```python
for sent,emb in zip(sentences,embeddings): #to get iterator based on both sentences, embeddings
    print('Sentence:',sent)
    print('Len(Embeddings):',len(emb))
    print('Embeddings:',emb)
    print()
```

```python
cosine_scores=util.cos_sim(embeddings,embeddings) #check within the embeddings, can pass different embeddings also
```

```python
#can use paraphrase mining to find the top ranked matching document, when dealing with hundreds and thousands of documents
paraphrases=util.paraphrase_mining(model,sentences)
```

```python
for sim in paraphrases[0:10]: #top 10 matches between the documents
    score,i,j=sim
    print(sentences[i],'<>',sentences[j],'-->',score)
```
