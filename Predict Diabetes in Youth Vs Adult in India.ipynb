```python
from sklearn.preprocessing import OrdinalEncoder

categories_order = [
    ["No", "Yes"],
    ["Unknown","Type 1","Type 2"],
    ["Sedentary","Moderate","Active"],
    ["Unhealthy","Moderate","Healthy"],
    ["No", "Yes"],
    ["No", "Yes"],
    ["No", "Yes"],
    ["Unknown","Type 1","Type 2"]
]
columns_to_encode = ["Family_History_Diabetes","Parent_Diabetes_Type","Physical_Activity_Level","Dietary_Habits","Smoking","Alcohol_Consumption","Prediabetes","Diabetes_Type"]

for col in columns_to_encode:
    df[col] = df[col].fillna("Unknown")

encoder = OrdinalEncoder(categories=categories_order)
df[columns_to_encode] = encoder.fit_transform(df[columns_to_encode])
print(df)
```

```python
df = pd.get_dummies(df, columns=["Gender","Region"])
print(df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_columns = ["Age","Family_Income","Genetic_Risk_Score","BMI","Fast_Food_Intake","Fasting_Blood_Sugar","HbA1c","Cholesterol_Level","Sleep_Hours","Stress_Level","Screen_Time"]
df[num_columns] = scaler.fit_transform(df[num_columns])
print(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import tensorflow as tf
```

```python
X = df.drop(["ID","Diabetes_Type"], axis=1)
y = df["Diabetes_Type"]
```

```python
from sklearn.preprocessing import OneHotEncoder
y_numpy = y.values
encoder = OneHotEncoder()
y_onehot = encoder.fit_transform(y_numpy.reshape(-1, 1)) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size=0.2, random_state=123)
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, accuracy_score

# Ensure that y_train and y_test are dense (convert sparse matrix to dense array)
y_train_dense = y_train.toarray() if hasattr(y_train, 'toarray') else y_train
y_test_dense = y_test.toarray() if hasattr(y_test, 'toarray') else y_test

# Determine num_classes based on the labels (for categorical classification)
num_classes = len(set(y_train_dense.flatten()))  # Get the number of unique classes in the labels

# One-hot encode the labels (ensure they are dense)
y_train_one_hot = to_categorical(y_train_dense, num_classes=num_classes)
y_test_one_hot = to_categorical(y_test_dense, num_classes=num_classes)

# Check and reshape the labels (ensure shape is [n_samples, n_classes])
# If the labels have extra dimensions (e.g., [n_samples, n_classes, 1]), flatten them
y_train_one_hot = y_train_one_hot.reshape(-1, num_classes)
y_test_one_hot = y_test_one_hot.reshape(-1, num_classes)

# Define the model with an Input layer to avoid the warning
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train.shape[1],)),  # Fix: Use Input layer instead of passing input_shape directly
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(y_train_one_hot.shape[1], activation='softmax')  # Output layer for one-hot encoded labels
])

# Compile the model with categorical crossentropy
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',  # Use categorical crossentropy
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train, y_train_one_hot,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

# Make predictions
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # Get the predicted class labels
y_test_classes = np.argmax(y_test_one_hot, axis=1)  # Convert one-hot encoded y_test to class labels

# Ensure the shapes match for classification report
print(f"y_pred_classes shape: {y_pred_classes.shape}")
print(f"y_test_classes shape: {y_test_classes.shape}")
```
