```python
ds = pd.read_csv(r"C:\Users\dell\Desktop\MyDocs\Docs\MK\dataset.csv")
ds.head()
```

```python
#missing values
ds.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
ds["Target"] = label_encoder.fit_transform(ds["Target"])
ds.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr_matrix = ds.corr()

# Plot the heatmap for the correlation matrix
plt.figure(figsize=(30, 15))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.7)

# Show the plot
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Calculate the correlation of each numerical column with the "Target" column
target_corr = ds.corr()["Target"]
sorted_target_corr = target_corr.sort_values(ascending=False)
# Print the correlation values with respect to "Target"
print(sorted_target_corr)
```

```python
X = ds.iloc[:,:-1].values
X
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
y = ds.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=0)
```

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train,y_train)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
clf.fit(X_train,y_train)
```

```python
from tensorflow.keras import regularizers
import tensorflow as tf
# Initialising ANN
ann = tf.keras.models.Sequential()

# Adding layers

ann.add(tf.keras.layers.Dense(units=64, activation='relu'))
ann.add(tf.keras.layers.Dropout(0.2))
ann.add(tf.keras.layers.Dense(units=32, activation='relu'))
ann.add(tf.keras.layers.Dropout(0.2))
ann.add(tf.keras.layers.Dense(units=3, activation='softmax'))  # Use softmax for multi-class classification
```

```python
ann.compile(optimizer='adam',
            loss='sparse_categorical_crossentropy',  # For integer-encoded labels
            metrics=['accuracy'])
ann.fit(X_train,y_train,batch_size = 32,epochs=50)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
# accuracy_score(y_test,ann.predict(X_test))
y_pred_prob = ann.predict(X_test)

# Get the predicted class labels by selecting the class with the highest probability
y_pred = np.argmax(y_pred_prob, axis=1)

# Print the result
print(y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
y_pred_XGB = classifier.predict(X_test)
```

```python
y_pred_SVC = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
# accuracy_score(y_test,ann.predict(X_test))
y_pred_prob = ann.predict(X_test)

# Get the predicted class labels by selecting the class with the highest probability
y_pred_ANN = np.argmax(y_pred_prob, axis=1)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cf_XGB = confusion_matrix(y_test,y_pred_XGB)
cf_SVC = confusion_matrix(y_test,y_pred_SVC)
cf_ANN = confusion_matrix(y_test,y_pred_ANN)
```

```python
acc_XGB = accuracy_score(y_test,y_pred_XGB)
acc_SVC = accuracy_score(y_test,y_pred_SVC)
acc_ANN = accuracy_score(y_test,y_pred_ANN)
```

```python
cr_XGB = classification_report(y_test,y_pred_XGB)
cr_SVC = classification_report(y_test,y_pred_SVC)
cr_ANN = classification_report(y_test,y_pred_ANN)
```
