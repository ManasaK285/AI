```python
df['x_age'] = df['x_age'].fillna(df['x_age'].median())
```

```python
df['x_case'] = df['x_case'].map({0: 'Control', 1: 'Cancer'})
```

```python
df['exam_year'] = pd.to_datetime(df['exam_year'], format='%Y')
df['year'] = df['exam_year'].dt.year
```

```python
df
```

```python
df = pd.get_dummies(df,columns=['x_case'],drop_first=True)
```

```python
df
```

```python
x = df[['x_age', 'year']]
y = df['x_case_Control']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
y_test
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=8, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=8, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1))
```

```python
ann.compile(optimizer='adam', loss='mean_squared_error')
```

```python
ann.fit(x_train, y_train, batch_size= 64, epochs = 120)
```

```python
import numpy as np

y_pred = ann.predict(x_test)
y_test = np.array(y_test)

np.set_printoptions(precision=2)
result = np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), axis=1)

print(result)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
```
