```python
print(f"before: {df.shape}") #before removing duplicate datapoints
df = df.drop_duplicates()
print(f"after: {df.shape}") #after removing duplicate datapoints
```

```python
pd.options.display.max_rows = None # we use this to show all features null values
df.isnull().sum()
```

```python
X = df.drop(columns=60,axis =1)
y = df[60]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)
```

```python
model = LogisticRegression() #load the model
model.fit(X_train,y_train)
model = LogisticRegression() #load the model
model.fit(X_train,y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
print(f"Accuracy on training dataset: {training_data_accuracy*100}")
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
print(f"Accuracy on training dataset: {test_data_accuracy*100}")
```
