# Task#1: Understanding problem statement and Business Usecase

Fashion training set consists of 70,000 images divided into 60,000 training and 10,000 testing samples. Dataset sample consists of 28x28 grayscale image, associated with a label from 10 classes. 

The 10 classes are as follows:  
0 => T-shirt/top
1 => Trouser
2 => Pullover
3 => Dress
4 => Coat
5 => Sandal
6 => Shirt
7 => Sneaker
8 => Bag
9 => Ankle boot

Each image is 28 pixels in height and 28 pixels in width, for a total of 784 pixels in total. Each pixel has a single pixel-value associated with it, indicating the lightness or darkness of that pixel, with higher numbers meaning darker. This pixel-value is an integer between 0 and 255.

# Task#4: Training the Model

```python
X_train=training[:,1:]/255 #/255-> normalizing
y_train=training[:,0]
```

```python
X_test=testing[:,1:]/255
y_test=testing[:,0]
```

```python
#validation dataset to help the model to generalize, avoid any overfitting
from sklearn.model_selection import train_test_split
```

```python
X_train,X_validate,y_train,y_validate=train_test_split(X_train,y_train,test_size=0.2,random_state=12345)
```

```python
X_train.shape
```

```python
X_validate.shape
```

```python
y_train.shape
```

```python
y_validate.shape
```

```python
#to put all these data in array format, so re-form into 28,28,1 to feed it to CNN
X_train = X_train.reshape(X_train.shape[0], *(28, 28, 1))
X_test = X_test.reshape(X_test.shape[0], *(28, 28, 1))
X_validate = X_validate.reshape(X_validate.shape[0], *(28, 28, 1))
```

```python
X_train.shape
```

```python
X_validate.shape
```

```python
X_test.shape
```

```python
#training model using keras-> advanced CNNs
import keras
```

```python
from keras.models import Sequential #to build network in sequential form
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model=Sequential()
```

```python
cnn_model.add(Conv2D(32,3,3,input_shape=(28,28,1),activation='relu')) #feature detectors->32, dimensions->3,3
```

```python
#adding maxpooling layer
cnn_model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
#flattening the model
cnn_model.add(Flatten())
```

```python
#cnn_model.add(Dense(output_dim=32,activation='relu'))-> removing output_dim to avoid missing positional argument 'units' error
cnn_model.add(Dense(32,activation='relu'))
```

```python
#output layer
cnn_model.add(Dense(10,activation='sigmoid')) #10 dimensions, as only 10 categories
```

```python
#using adam optimizer
cnn_model.compile(loss ='sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.001),metrics =['accuracy'])
```

```python
epochs=50
```

```python
#not nb_epochs, just epochs
epochs = 50
history = cnn_model.fit(X_train,
                        y_train,
                        batch_size = 512,
                        epochs = epochs,
                        verbose = 1,
                        validation_data = (X_validate, y_validate))
```

# Task#5: Evaluating the Model

```python
evaluation = cnn_model.evaluate(X_test, y_test)
print('Test Accuracy : {:.3f}'.format(evaluation[1]))
```

```python
# get the predictions for the test data
predicted_classes = np.argmax(cnn_model.predict(X_test),axis=1)
```

```python
predicted_classes
```

```python
L = 5
W = 5
fig, axes = plt.subplots(L, W, figsize = (12,12))
axes = axes.ravel() # 

for i in np.arange(0, L * W):  
    axes[i].imshow(X_test[i].reshape(28,28))
    axes[i].set_title("Prediction Class = {:0.1f}\n True Class = {:0.1f}".format(predicted_classes[i], y_test[i]))
    axes[i].axis('off')

plt.subplots_adjust(wspace=0.5)
```

```python
#plotting confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_classes)
plt.figure(figsize = (14,10))
sns.heatmap(cm, annot=True)
```

```python
#printing classification report
from sklearn.metrics import classification_report

num_classes = 10
target_names = ["Class {}".format(i) for i in range(num_classes)]

print(classification_report(y_test, predicted_classes, target_names = target_names))
```

# Task#6: Improving Model

- By increasing the number of kernels (64 instead of 32)or
- Dropouts-> during training, take some of samples, disconnect them to have a better training, reducing the co-dependencies between the neurons(Regularization Technique), avoiding overfitting and making the model more generic

```python
#increasing kernel size
cnn_model2 = Sequential()

#increasing to 128 from 32 and adding Dropout, with dropout, accuracy was just 84 so commenting and re-training, 64-87% accuracy
cnn_model2.add(Conv2D(128,3,3,input_shape=(28,28,1),activation='relu'))
cnn_model2.add(MaxPooling2D((2,2), padding='same'))

#cnn_model2.add(Dropout(0.25))

cnn_model2.add(Conv2D(128,3, 3, activation='relu'))
cnn_model2.add(MaxPooling2D((2,2), padding='same'))

cnn_model2.add(Flatten())
cnn_model2.add(Dense(128, activation = 'relu'))
cnn_model2.add(Dense(10, activation = 'sigmoid'))
```

```python
cnn_model2.compile(loss ='sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.001),metrics =['accuracy'])
```

```python
epochs2 = 50
history2 = cnn_model2.fit(X_train,
                        y_train,
                        batch_size = 512,
                        epochs = epochs2,
                        verbose = 1,
                        validation_data = (X_validate, y_validate))
```

# Re-Evaluating the new Model

```python
evaluation2 = cnn_model2.evaluate(X_test, y_test)
print('Test Accuracy : {:.3f}'.format(evaluation[1]))
```
