# Step#4: Creating Training and Testing Data

```python
IceCream
```

```python
X=IceCream['Temperature']
```

```python
y=IceCream['Revenue']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

# Step#5: Train the Model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression(fit_intercept=True) #obtaining x and y intercept
```

```python
regressor.fit(pd.DataFrame(X_train),pd.DataFrame(y_train))
```

```python
print('Linear Model Coefficient (m)',regressor.coef_)
print('Linear Model Intercept (b)',regressor.intercept_)
```

```python
#using fit_intercept as false
from sklearn.linear_model import LinearRegression
regressor=LinearRegression(fit_intercept=False) #obtaining x and y intercept
regressor.fit(pd.DataFrame(X_train),pd.DataFrame(y_train))
```

```python
print('Linear Model Coefficient (m)',regressor.coef_)
print('Linear Model Intercept (b)',regressor.intercept_)
```

# Step#6: Test the Model

```python
X_test.shape
```

```python
X_test
```

```python
y_predict=regressor.predict(pd.DataFrame(X_test))
```

```python
y_predict
```

```python
plt.scatter(X_train, y_train, color = 'gray')
plt.plot(X_train, regressor.predict(pd.DataFrame(X_train)), color = 'red')
plt.ylabel('Revenue [dollars]')
plt.xlabel('Temperature [degC]')
plt.title('Revenue Generated vs. Temperature @Ice Cream Stand(Training dataset)')
```

```python
#plotting using testing data
plt.scatter(X_test, y_test, color = 'blue')
plt.plot(X_test, regressor.predict(pd.DataFrame(X_test)), color = 'red')
plt.ylabel('Revenue [dollars]')
plt.xlabel('Temperature [degC]')
plt.title('Revenue Generated vs. Temperature @Ice Cream Stand(Training dataset)')
```
