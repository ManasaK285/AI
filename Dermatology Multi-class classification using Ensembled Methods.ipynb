```python
data['age']=pd.DataFrame(data['age'])

print(data['age'])

data['age']=pd.to_numeric(data['age'], errors='coerce')

data['age'].dtypes
```

```python
x=data.drop(['age','class'],axis=1)

y=data['class']

from sklearn.preprocessing import MinMaxScaler,LabelEncoder

X=MinMaxScaler().fit_transform(x)

X=pd.DataFrame(X).join(data['age'])

y=LabelEncoder().fit_transform(y)
```

```python
pd.DataFrame(X)
```

```python
X=X.dropna(axis=1)

X=X.dropna(axis=0)
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

import numpy as np

xtrain,xtest,ytrain,ytest=train_test_split(np.array(X),np.array(y),test_size=0.3,random_state=43)
```

```python
xtrain[:5],ytrain[:5]
```

```python
def groundTruth(arr):

    FP = (arr.sum(axis=0) - np.diag(arr)).astype(float).sum()

    FN = (arr.sum(axis=1) - np.diag(arr)).astype(float).sum()

    TP = (np.diag(arr)).astype(float).sum()

    TN = (arr.sum() - (FP + FN + TP)).astype(float).sum()

  # recall

    TPR = (TP/(TP+FN))

  # precision

    precision = (TP/(TP+FP))

  # accuracy

    accuracy = ((TP+TN)/(TP+FP+FN+TN))

    f1score=(2*precision*TPR)/(precision+TPR)



    return "Precision : "+str(round(precision*100,2))+"\nRecall : "+str(round(TPR*100,2))+"\nAccuracy : "+str(round(accuracy*100,2))+"\nF1 Score : "+str(round(f1score*100,2))+"\n"
```

```python
from sklearn.tree import DecisionTreeClassifier

modelDT=DecisionTreeClassifier(max_depth=8).fit(xtrain,ytrain)

predictionDT=modelDT.predict(xtest)

print(f"Score : {modelDT.score(xtest,ytest)}")
```

```python
from sklearn.naive_bayes import GaussianNB

modelNB=GaussianNB().fit(xtrain,ytrain)

predictionNB=modelNB.predict(xtest)

print(f"Score : {modelNB.score(xtest,ytest)}")
```

```python
from sklearn import svm

modelSVM=svm.SVC(kernel='poly', degree=6, C=1).fit(xtrain,ytrain)

predictionSVM=modelSVM.predict(xtest)

print(f"Score : {modelSVM.score(xtest,ytest)}")
```

```python
from sklearn.metrics import confusion_matrix,roc_curve

fig,ax=plt.subplots(1,3,figsize=(15,3))

ax[0].set_title("Decision Tree")

snb.heatmap(confusion_matrix(ytest,predictionDT),ax=ax[0],cmap="YlOrBr", annot=True)

ax[0]=plt.figtext(0.125,1,groundTruth(confusion_matrix(ytest,predictionDT)))



ax[1].set_title("Naive Bayes")

snb.heatmap(confusion_matrix(ytest,predictionNB),ax=ax[1],cmap="GnBu", annot=True)

ax[1]=plt.figtext(0.4,1,groundTruth(confusion_matrix(ytest,predictionNB)))



ax[2].set_title("SVM (Polynomial Kernel)")

snb.heatmap(confusion_matrix(ytest,predictionSVM),ax=ax[2],cmap="YlGn", annot=True)

ax[2]=plt.figtext(0.675,1,groundTruth(confusion_matrix(ytest,predictionSVM)))
```

```python
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier



modelABC=AdaBoostClassifier(n_estimators = 500, learning_rate = 0.5).fit(xtrain,ytrain)

predictionABC=modelABC.predict(xtest)

print(f"Score : {modelABC.score(xtest,ytest)}")
```

```python
modelRF=RandomForestClassifier(max_depth=10).fit(xtrain,ytrain)

predictionRF=modelRF.predict(xtest)

print(f"Score : {modelRF.score(xtest,ytest)}")
```

```python
from xgboost import XGBClassifier

modelXB = XGBClassifier(n_estimators = 500, learning_rate = 0.001).fit(xtrain,ytrain)

predictionXB=modelXB.predict(xtest)

print(f"Score : {modelXB.score(xtest,ytest)}")
```

```python
from sklearn.metrics import confusion_matrix

fig,ax=plt.subplots(1,3,figsize=(15,3))

ax[0].set_title("AdaBoost")

snb.heatmap(confusion_matrix(ytest,predictionABC),ax=ax[0],cmap="cividis", annot=True)

ax[0]=plt.figtext(0.125,1,groundTruth(confusion_matrix(ytest,predictionABC)))



ax[1].set_title("Random Forest")

snb.heatmap(confusion_matrix(ytest,predictionRF),ax=ax[1],cmap="summer", annot=True)

ax[1]=plt.figtext(0.4,1,groundTruth(confusion_matrix(ytest,predictionRF)))



ax[2].set_title("XBoost")

snb.heatmap(confusion_matrix(ytest,predictionXB),ax=ax[2],cmap="copper", annot=True)

ax[2]=plt.figtext(0.675,1,groundTruth(confusion_matrix(ytest,predictionXB)))
```
