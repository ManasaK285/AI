# Step#5: Create Training and Testing Data

```python
X=fuel_economy_df[['Horse Power']] #df[[]] 2 []
```

```python
y=fuel_economy_df[['Fuel Economy (MPG)']]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Step#5: Training the Model

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression(fit_intercept=True) #include intercept also, if false, passes through origin
```

```python
regressor.fit(X_train,y_train)
```

```python
print('Linear Model Coeff (m):', regressor.coef_)
print('Linear Model Coeff (b):', regressor.intercept_)
```

```python
# checking for intercept=False
from sklearn.linear_model import LinearRegression
regressor=LinearRegression(fit_intercept=False) #include intercept also, if false, passes through origin
regressor.fit(X_train,y_train)
```

```python
print('Linear Model Coeff (m):', regressor.coef_)
print('Linear Model Coeff (b):', regressor.intercept_) #with intercept=False, slope became positive, which ideally isn't
```

# Step#6: Test the Model

```python
y_predict=regressor.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
#plotting training data - 70 samples
plt.scatter(X_train, y_train, color = 'orange')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.xlabel('Horse Power (HP)')
plt.ylabel('MPG')
plt.title('HP vs. MPG (Training Set)')
```

```python
#plotting the test data--> shows 30 samples
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_test, regressor.predict(X_test), color = 'blue')
plt.xlabel('Horse Power (HP)')
plt.ylabel('MPG')
plt.title('HP vs. MPG (Training Set)')
```
