```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
data["Engagement_Level"]=label.fit_transform(data["Engagement_Level"])
data.drop(columns="User_ID", inplace=True)
print("Mapping:", dict(zip(label.classes_, label.transform(label.classes_))))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Generate the heatmap
plt.figure(figsize=(18,12))
sns.heatmap(numeric_data.corr(), annot=True, cmap="viridis")
plt.title("Correlation Matrix",fontsize=30)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x=data.drop(columns="Purchase_Likelihood")
y=data["Purchase_Likelihood"]
x_test,x_train, y_test,y_train=train_test_split(x,y)

model=LogisticRegression(random_state=42,solver="liblinear")
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print("Accuracy of Logistic Regression is", acc)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
conf_mat=confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat,annot=True,cmap="viridis",fmt="d")
plt.title("Confusion Matrix",fontsize=20)
plt.show()
```

```python
from sklearn.svm import SVC

# Create and train the SVM model
model = SVC()
model.fit(x_train, y_train)

# Predict the labels for the test set
y_pred = model.predict(x_test)

# Calculate the accuracy of the model
accuracy = model.score(x_test, y_test)

# Print the accuracy
print("Accuracy of SVC is",accuracy)
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model to the data
kmeans.fit(data)

# Get the cluster labels (assigned clusters for each data point)
labels = kmeans.labels_

# Add the labels to the data as a new column
df = pd.DataFrame(data)  
df['Cluster'] = labels  

# Sort the DataFrame by the 'Cluster' column
df_sorted = df.sort_values(by='Cluster')

# Perform PCA to reduce the dimensionality to 2 for visualization
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data)

# Plot the data points in 2D with cluster colors
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=labels, cmap='viridis')

plt.title("K-Means Clustering")
plt.show()
```
